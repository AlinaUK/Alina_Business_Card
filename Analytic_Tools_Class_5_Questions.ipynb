# **PART 2: Using Polynomial Regression**

**Question 1.1:** Load the *Years_salary_data.csv* file in the Colab and *read* it as a dataframe.

Print and examine the dataset.

```python
# Provide your solution here
import pandas as pd

data_ys = pd.read_csv('Years_salary_data.csv')
data_ys.head()
```

**Question 1.2:** Let's plot a scatterplot to examine the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Plotting the scatterplot
sns.scatterplot(data=data_ys, x='Years', y='Salary')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Polynomial Regression")
plt.show()
```

**Question 2:** Plot the years-salary data using a lmplot. Use in **x-axis** the **Years** and as **y-axis** the **Salary**.

```python
# Provide your solution here
sns.lmplot(x='Years', y='Salary', data =data_ys, height = 10)
plt.show()
```

**Question 3:** Create the **X (Years)** and **y (Salary)** variables of your dataset.

```python
# Provide your solution here

X= data_ys['Years']
y= data_ys['Salary']
```

**Question 4:** Create the Linear Regression model. Fit **X** and **y**.


```python
# Provide your solution here
X = data_ys[['Years']]  # Double brackets make it a DataFrame (2D)
y = data_ys['Salary']

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X, y)
```

**Question 5:** Create the Polynomial Regression model (degree=8). Fit X and y.

Study and run the following code.

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=8)
X_poly=poly.fit_transform(X)
poly.fit(X_poly,y)
linreg2=LinearRegression()
linreg2.fit(X_poly,y)

```

**Question 6:** Plot the Linear Regression model.

 Study and run the following code. Is the Regression line a good fit?

```python
# Visualising the Linear Regression results

plt.scatter(X, y, color = 'blue')

plt.plot(X, linreg.predict(X), color = 'red')
plt.title('Linear Regression')
plt.xlabel('Year')
plt.ylabel('Salary')

plt.show()

# The regression line does not represent the data
```

**Question 7:** Plot the Polynomial Regression results.

Study and run the following code. Is the Polynomial line a good fit?

```python
# Visualising the Polynomial Regression results
plt.scatter(X,y, color = 'blue')
plt.plot(X, linreg2.predict(poly.fit_transform(X)), color='red')
plt.title('Polynomial Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()



# The Polynomial line represent better the data!
```

**Question 8:** Let us change the degree of the Polynomial line.

Study and run the following code using the following.

*   degree=4
*   degree=10

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=4)
X_poly=poly.fit_transform(X)
poly.fit(X_poly,y)
linreg2=LinearRegression()
linreg2.fit(X_poly,y)



# Visualising the Polynomial Regression results

```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 10)
X_poly = poly.fit_transform(X)

poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'blue')

plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'red')
plt.title('Polynomial Regression')
```

> The degree affects the fit of our Polynomial line, if the degree is very small our model does not fit well to the data, if the degree is very high the model overfits the data. We need to find a balance to control it.

**Question 9:** Create a Polynomial Regression model using degree 11. Use the code from the previous examples.

```python
# Provide your solution here
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=11)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
```

**Question 10:** Predict the salary for years of experience as follows.
*   32.5
*   20.5

Create a linear and Polynomial Regression prediction.

Study and run the following code.

Which model is more accurate?


```python
# Predicting a new result with Linear Regression
from sklearn.linear_model import LinearRegression
linreg.predict([[3.25]])

# Predicting a new result with Polymonial Regression
print(pol_reg.predict(poly_reg.fit_transform([[20.5]])))

print("Linear Regression Prediction for 32.5 years:", linreg.predict([[32.5]]))
print("Polynomial Regression Prediction for 32.5 years:", pol_reg.predict(poly_reg.fit_transform([[32.5]])))
print("Linear Regression Prediction for 20.5 years:", linreg.predict([[20.5]]))
print("Polynomial Regression Prediction for 20.5 years:", pol_reg.predict(poly_reg.fit_transform([[20.5]])))
# Answer: __Polynomial Regression is a better fit model in this dataset___
```

**Question 3:** We will need to split our dataset into feature data (X) and target data (y).

We will use the **sepal length** and **width**, and the petal **length** and **width** to predict a flower's class (variety).

So, the feature data is:
*  sepal.length
*  sepal.width
*  petal.length
*  petal.width

The target data is:
*  variety

Extract data for the feature variables (and associated columns) and store it in a new dataframe called *iris_data_feature_data*. This will be our **X** variable.

Also, create the *iris_data_feature_target* dataframe. The new dataframe will store only the target values that is the variety column. This will be our **y** variable.


```python
# Provide your solution here
iris_data_feature_data = iris_data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
X = iris_data_feature_data
iris_data_target_data = iris_data['variety']
y = iris_data_target_data
# Display the first 5 rows of independent data
print(X.head())


```

**Question 4:** Create a new *LogisticRegression model* and then fit the **X** and **y** dataframes that we just created. Make sure you revise the class slides to create your model.

```python
# Provide your solution here
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(max_iter=200)
logreg.fit(X,np.ravel(y))


```

**Question 5:** Create the *y_pred* variable to store the predicted values of ***X*** data.

```python
# Provide your solution here
y_pred = logreg.predict(X)
y_pred
```

**Question 6:** Use the *metrics.accuracy_score* method to predict the accuracy score, between ***y*** and ***y_pred***.

```python
# Provide your solution here

from sklearn import metrics
metrics.accuracy_score(y, y_pred)
```

**Question 8:** Provide a script to create a KNN classifier for *k=5*.

* Then fit the model using the X and y data.
* Create the *y_pred_knn5* to make predictions using the X data.
* Finally print the *accuracy_score* for the *y* and *y_pred_knn5*.

Round the accuracy score to only two decimals. What is the accuracy score?


```python
# Provide your solution here
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X, np.ravel(y))
y_pred_knn5 = knn.predict(X)
round(metrics.accuracy_score(y, y_pred_knn5), 2)




# Accuracy is: ___0.97___
```
