In our case we will map each value to a number, such as:

* 'Overcast': 0, 'Rainy': 1, 'Sunny': 2
* 'Cool': 0, 'Hot': 1, 'Mild': 2

The **preprocessing package** can help us to transform the data.

Study and run the following script. The function from question 10, is just a way to print the mappings, e.g. Play No: 0, Play Yes: 1. Keep this as a reference for the future on how to extract the mappings.

**Question 1:** Let us explore various metrics of the linear regression models.

In this section we will just explore how to mesure **MSE**, **MAE** and **RMSE** scores. Feel free to use this in your coursework.

```python
# Provide your solution here
import numpy as np

# Create X and y
X = iris_data[['sepal.width','petal.length','petal.width']]
y = iris_data[['sepal.length']]

# Create the Linear Regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

# Fit the model
linreg.fit(X, y)

# Create the predicted variable using the y data
y_pred = linreg.predict(X)
```

**Question 2:** Let's create the **MSE, MAE** and **RMSE** scores.

Study and run the following.

```python
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

MSE = mean_squared_error(y, y_pred)
MAE = mean_absolute_error(y, y_pred)
RMSE = math.sqrt(MSE)

print("MSE: ", round(MSE,2))
print("MAE: ", round(MAE,2))
print("RMSE: ", round(RMSE,2))
```

**Question 12:** Study and run the following code.
What are the best k values?

```python
# Provide your solution here
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)

k_range = range(1,30)

scores = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, np.ravel(y_train))
  y_pred = knn.predict(X_test)
  accuracy = metrics.accuracy_score(y_test, y_pred)
  scores.append(accuracy)


print(scores)

plt.plot(k_range, scores)
plt.xlabel("Value of k for KNN")
plt.ylabel("Testing accuracy")

# The best k values are: 6-17 and 19
```
